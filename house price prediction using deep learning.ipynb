```python
import pandas as pd
import matplotlib.pyplot as  plt
import tensorflow as tf
import numpy as np
```

```python
data=pd.read_csv("data.csv")
data.head()
```

```python
data.columns
```

```python
data=data[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above']]
data.head()
```

```python
X=data.iloc[:,1:]
Y=data[['price']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X1=scaler.fit_transform(X)
```

```python
from tensorflow import keras
```

```python
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=(9,))
                       ])
```

```python

model.compile(optimizer='sgd',loss='mean_squared_error')
```

```python
model.fit(X1,Y,epochs=500)
```
